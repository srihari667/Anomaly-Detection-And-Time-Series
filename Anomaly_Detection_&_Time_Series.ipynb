1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

    **Anomaly Detection** is the process of identifying **data points, events, or observations that deviate significantly** from the normal pattern or expected behavior within a dataset.
    These unusual observations are called **anomalies** or **outliers**.

    It is widely used in:

    * **Fraud detection** (credit card transactions)
    * **Network security** (intrusion detection)
    * **Manufacturing** (fault detection)
    * **Healthcare** (disease outbreak detection)
    * **Finance** (stock price irregularities)


    ### **Types of Anomalies**

    There are **three main types** of anomalies:


    #### **1. Point Anomalies (Global Outliers)**

    * **Definition:**
      A single data point is considered anomalous if it is **too far** from the rest of the data in the same context.

    * **Example:**
      Suppose the average daily temperature in a city during winter is between **10°C and 20°C**, but one day the temperature suddenly jumps to **40°C**.
      This single data point (40°C) is a **point anomaly**.

    * **Use Case Example:**

      * A sudden large withdrawal from a bank account (possible fraud).
      * A server showing 100% CPU usage when it usually stays below 40%.


    #### **2. Contextual Anomalies (Conditional Anomalies)**

    * **Definition:**
      A data point is anomalous **only in a specific context**.
      The same value may be normal in one situation but abnormal in another.

    * **Example:**
      A temperature of **30°C** is normal in **summer**, but **anomalous in winter**.
      The “context” here is the **season (time)**.

    * **Use Case Example:**

      * A sudden drop in website traffic during the night may be normal,
        but the same drop during business hours might be suspicious.
      * Detecting abnormal heart rate given a person’s activity level (resting vs running).


    #### **3. Collective Anomalies**

    * **Definition:**
      A group or sequence of data points is considered anomalous when viewed **together**, even if individual points appear normal.

    * **Example:**
      In a time series of network traffic, a sudden burst of many small packets might indicate a **DDoS attack** — the individual packets aren’t strange, but their **collective behavior** is.

    * **Use Case Example:**

      * Multiple failed login attempts within a short time period (possible hacking attempt).
      * A group of consecutive readings from a machine sensor showing a pattern that indicates malfunction.

2.  Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.



    ### **1. Isolation Forest (IF)**

    **Approach:**

    * Based on **ensemble learning** using decision trees.
    * It works by **randomly partitioning** the data — anomalies are easier to isolate (require fewer splits), while normal points need more splits.
    * The **average path length** to isolate a point determines how “anomalous” it is.

    **Key Idea:**
    Anomalies are “few and different,” so they are isolated quickly.

    **Strengths:**

    * Scales well to **large, high-dimensional datasets**.
    * Works efficiently with **continuous** and **numerical data**.
    * Handles **non-linear relationships** well.

    **Weaknesses:**

    * Not ideal for **clusters with varying densities**.
    * Results may vary for **small datasets**.

    **Use Cases:**

    * Detecting **fraudulent transactions** in large datasets.
    * Identifying **server/network intrusions**.
    * Spotting **defective parts** in manufacturing data.


    ### **2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

    **Approach:**

    * A **density-based clustering** algorithm that groups together closely packed points and marks low-density points as **noise/outliers**.
    * Controlled by two parameters:

      * **ε (eps):** Neighborhood radius.
      * **minPts:** Minimum points to form a dense region.

    **Key Idea:**
    Outliers lie in **low-density regions** not belonging to any cluster.

    **Strengths:**

    * Finds **clusters of arbitrary shape**.
    * Automatically detects **noise/outliers**.
    * No need to predefine the number of clusters.

    **Weaknesses:**

    * Sensitive to **parameter selection (ε and minPts)**.
    * Struggles with **varying densities**.
    * Doesn’t scale well for **high-dimensional data**.

    **Use Cases:**

    * **Spatial data analysis** (e.g., detecting anomalies in GPS or geolocation data).
    * **Customer segmentation** where natural clusters exist.
    * **Image segmentation** and **pattern recognition**.


    ### **3. Local Outlier Factor (LOF)**

    **Approach:**

    * A **density-based anomaly detection** algorithm.
    * Compares the **local density** of each point with its neighbors to compute a **“local outlier factor”** score.
    * A point with much lower density than its neighbors is an **outlier**.

    **Key Idea:**
    Anomalies are **locally sparse** compared to their surrounding neighborhood.

    **Strengths:**

    * Effective for **datasets with varying densities**.
    * Detects **local anomalies** that global methods might miss.

    **Weaknesses:**

    * Sensitive to **number of neighbors (k)** parameter.
    * **Not scalable** to very large datasets.
    * Performance drops in **high dimensions**.

    **Use Cases:**

    * **Credit card fraud detection** where outliers depend on local context.
    * **Network traffic anomaly detection**.
    * **Sensor or IoT data** with variable density regions.


3. What are the key components of a Time Series? Explain each with one example.


    ###  **1. Trend (T)**

    **Definition:**
     A long-term **upward or downward movement** in the data over a period of time.
    It shows the **overall direction** of the series but ignores short-term fluctuations.

    **Example:**
     The **annual average temperature** of Earth increasing over the last 50 years shows an **upward trend** due to climate change.


    ###  **2. Seasonality (S)**

    **Definition:**
    A **repeating pattern or fluctuation** that occurs at **regular intervals** due to **seasonal factors** like weather, holidays, or business cycles.

    **Example:**
      Retail sales increase every **December** due to holiday shopping — this is a **seasonal pattern** repeating yearly.


    ###  **3. Cyclic Component (C)**

    **Definition:**
    Represents **long-term oscillations** in the data that occur over **irregular time intervals**, often linked to **economic or business cycles**.
    Unlike seasonality, **cycle lengths are not fixed**.

    **Example:**
      The **stock market** experiences **boom and recession** phases every few years, forming **cyclic patterns**.


    ###  **4. Irregular or Random Component (R)**

    **Definition:**
    Represents **unpredictable or random variations** caused by unexpected events (natural disasters, strikes, pandemics, etc.).
    These fluctuations **cannot be predicted or modeled** easily.

    **Example:**
      A **sudden drop in tourism** due to a **pandemic or earthquake** is an irregular component.

4.  Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?


    A **time series is said to be stationary** if its **statistical properties** — such as **mean**, **variance**, and **autocorrelation** — remain **constant over time**.
    In other words, the behavior of the series does **not depend on the time** at which you observe it.

    ### **Key Characteristics of a Stationary Series**

    | Property              | Stationary Series               | Non-Stationary Series     |
    | --------------------- | ------------------------------- | ------------------------- |
    | **Mean**              | Constant over time              | Changes with time (trend) |
    | **Variance**          | Constant (no volatility change) | Varies over time          |
    | **Covariance**        | Depends only on lag             | Depends on time and lag   |
    | **Trend/Seasonality** | Absent                          | Present                   |


    ### **Example**

    * **Stationary Example:** Daily temperature **deviations** (after removing trend/seasonal effects).
    * **Non-Stationary Example:** **Monthly sales** that increase steadily due to business growth (trend present).


    ### **Why Stationarity Matters**

    Most **time series forecasting models** (like ARIMA, SARIMA) **require stationarity** because:

    * The statistical relationships are stable.
    * It ensures **consistent model parameters** over time.
    * Improves **forecast accuracy**.


    ### **Testing for Stationarity**

    #### 1. **Visual Inspection**

    * Plot the time series.
    * Look for a **trend**, **seasonal patterns**, or **changing variance**.
    * Example: A steadily rising plot usually means **non-stationary**.

    #### 2. **Summary Statistics**

    * Divide the series into two halves.
    * Compare **mean** and **variance** — if they’re significantly different → non-stationary.

    #### 3. **Statistical Tests**

    The most commonly used test is the **Augmented Dickey–Fuller (ADF) Test**.

    * **Null Hypothesis (H₀):** The series is **non-stationary** (unit root exists).
    * **Alternative Hypothesis (H₁):** The series is **stationary**.

      **If p-value < 0.05 → Reject H₀ → Stationary**
      **If p-value > 0.05 → Fail to reject H₀ → Non-stationary**


    ### **Transforming a Non-Stationary Series**

    #### 1. **Differencing**

    Subtract the previous observation from the current one:
    [
    Y'*t = Y_t - Y*{t-1}
    ]

    * Removes **trend** and stabilizes mean.
    * Can apply **first-order** or **second-order differencing** depending on data.

    #### 2. **Log Transformation**

    Apply log to stabilize **variance**:
    [
    Y'_t = \log(Y_t)
    ]

    * Useful when fluctuations increase with the level of the series.

    #### 3. **Detrending**

    * Fit a **regression line** and subtract the trend component.
    * Removes **systematic increase or decrease** in the mean.

    #### 4. **Seasonal Differencing**

    Subtract the value from the same season of the previous cycle:
    [
    Y'*t = Y_t - Y*{t-s}
    ]
    (where *s* = seasonal period, e.g., 12 for monthly data with yearly seasonality)


5.  Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.



    ##  **1. AR (AutoRegressive Model)**

    ### **Structure:**

    * Uses **past values** of the variable itself to predict the future.
    * The model assumes that the **current value** depends linearly on its **previous p values**.

    [
    Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + ... + \phi_p Y_{t-p} + \epsilon_t
    ]

    ### **Key Parameters:**

    * **p:** number of lag observations (autoregressive terms)

    ### **When to Use:**

    * When there is a **strong correlation** between a variable and its **past values**.
    * No trend or seasonality.

    ### **Example:**

    Predicting **stock prices** using their **past few days’ values**.


    ##  **2. MA (Moving Average Model)**

    ### **Structure:**

    * Uses **past forecast errors (residuals)** to predict the current value.
    * The model smooths out short-term fluctuations.

    [
    Y_t = c + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + ... + \theta_q \epsilon_{t-q}
    ]

    ### **Key Parameters:**

    * **q:** number of lagged forecast errors

    ### **When to Use:**

    * When data shows **random shocks or noise** but no clear trend or seasonality.

    ### **Example:**

    Forecasting **daily demand fluctuations** affected by short-term random events.


    ##  **3. ARIMA (AutoRegressive Integrated Moving Average)**

    ### **Structure:**

    * Combines **AR** and **MA** models with **differencing (I)** to handle **non-stationary** data.

    [
    ARIMA(p, d, q)
    ]

    Where:

    * **p** = AR terms
    * **d** = number of differencing operations
    * **q** = MA terms

    [
    Y'*t = c + \phi_1 Y'*{t-1} + ... + \theta_1 \epsilon_{t-1} + ... + \epsilon_t
    ]

    ### **When to Use:**

    * When the series shows a **trend but no seasonality**.
    * Works well for **short-term forecasting**.

    ### **Example:**

    Forecasting **monthly sales** after removing trend (no seasonal pattern).


    ##  **4. SARIMA (Seasonal ARIMA)**

    ### **Structure:**

    * Extends ARIMA to include **seasonal components**.
    * Suitable for data with **repeating seasonal patterns** (e.g., yearly, monthly).

    [
    SARIMA(p, d, q)(P, D, Q)_s
    ]

    Where:

    * **p, d, q:** Non-seasonal ARIMA parameters
    * **P, D, Q:** Seasonal ARIMA parameters
    * **s:** Seasonal period (e.g., 12 for monthly data with yearly seasonality)

    ### **When to Use:**

    * When time series shows **trend + seasonality**.

    ### **Example:**

    Forecasting **monthly airline passengers** where the number of passengers peaks every year.


    ##  **5. SARIMAX (Seasonal ARIMA with Exogenous Variables)**

    ### **Structure:**

    * Builds on SARIMA but includes **external (exogenous) variables** that may influence the series.

    [
    Y_t = SARIMA(p, d, q)(P, D, Q)_s + \beta X_t + \epsilon_t
    ]

    Where:

    * ( X_t ) = external variable(s) (e.g., marketing spend, temperature)

    ### **When to Use:**

    * When the target variable depends on **its own history** and also **other external factors**.

    ### **Example:**

    Predicting **energy consumption** based on **past usage (time factor)** and **temperature (exogenous variable)**.

```python
#7.  Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

data = pd.read_csv("NYC_taxi_fare_data.csv")

num_cols = ["fare_amount", "trip_distance", "tip_amount", "tolls_amount", "total_amount"]

num_data = data[num_cols].dropna()

aligned_data = data.loc[num_data.index]

iso_forest = IsolationForest(contamination=0.02, random_state=42)
aligned_data["anomaly"] = iso_forest.fit_predict(num_data)

anomalies = aligned_data[aligned_data["anomaly"] == -1]
normals = aligned_data[aligned_data["anomaly"] == 1]

plt.figure(figsize=(10, 6))
plt.scatter(normals["fare_amount"], normals["total_amount"], c="blue", label="Normal", alpha=0.5)
plt.scatter(anomalies["fare_amount"], anomalies["total_amount"], c="red", label="Anomaly", alpha=0.7)
plt.title("Isolation Forest - NYC Taxi Fare Anomaly Detection")
plt.xlabel("Fare Amount ($)")
plt.ylabel("Total Amount ($)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Total records used: {len(aligned_data)}")
print(f"Detected anomalies: {len(anomalies)}")
```

```python
#8.  Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results


import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = pd.read_csv("AirPassengers.csv")
data.columns = ["Month", "Passengers"]

data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)

data = data.asfreq('MS')

plt.figure(figsize=(10, 5))
plt.plot(data, label="Original Data", color="blue")
plt.title("Monthly Air Passengers (1949–1960)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)
plt.show()

model = SARIMAX(data["Passengers"],
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

sarima_model = model.fit(disp=False)

forecast_steps = 12
forecast = sarima_model.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1),
                               periods=forecast_steps, freq='MS')

forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

forecast_df = pd.DataFrame({
    "Forecast": forecast_mean.values
}, index=forecast_index)

plt.figure(figsize=(10, 6))
plt.plot(data.index, data["Passengers"], label="Historical Data", color="blue")
plt.plot(forecast_df.index, forecast_df["Forecast"], label="Forecast (Next 12 Months)", color="red")
plt.fill_between(forecast_index,
                 conf_int.iloc[:, 0],
                 conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title("SARIMA Forecast - AirPassengers Dataset")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)
plt.show()

print("\n Forecasted Passenger Numbers (Next 12 Months):")
print(forecast_df)
```

```python
#9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

air = pd.read_csv('AirPassengers.csv')

air.columns = air.columns.str.strip().str.replace('#','').str.replace(' ','')

air.rename(columns={'Passengers':'Passengers'}, inplace=True)
air.rename(columns={'Month':'Month'}, inplace=True)

air['Month'] = pd.to_datetime(air['Month'])
air.set_index('Month', inplace=True)

X = air['Passengers'].values.reshape(-1,1)

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)
lof_scores = -lof.negative_outlier_factor_

air['LOF_Pred'] = y_pred
air['LOF_Score'] = lof_scores

plt.figure(figsize=(12,6))
plt.plot(air.index, air['Passengers'], color='blue', label='Passengers')

outliers = air[air['LOF_Pred']==-1]
plt.scatter(outliers.index, outliers['Passengers'], color='red', label='Anomalies', s=50)

plt.title('AirPassengers - LOF Anomaly Detection')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()

print("Detected anomalies:")
print(outliers)
```

10. You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:

● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
● How would you validate and monitor the performance over time?
● How would this solution help business decisions or operations?


## ** Detecting Anomalies in Streaming Data**

The dataset is **high-frequency (15-min intervals)** with features like **timestamp, region, weather, and energy usage**.

### **Approach**

1. **Feature Engineering**

   * Create features from timestamp: `hour`, `day_of_week`, `is_holiday`.
   * Include **weather features**: temperature, humidity, wind, etc.
   * Use **rolling statistics**: mean, std deviation over past N intervals.

2. **Choose Anomaly Detection Algorithm**

   * **Isolation Forest (IF)**  Best for streaming & high-dimensional data.

     * Pros: Fast, handles large datasets, detects both global and local anomalies.
   * **Local Outlier Factor (LOF)**

     * Good for offline detection but not ideal for **real-time streaming** because it needs the full dataset for local density computation.
   * **DBSCAN**

     * Works well for spatial clustering but **not efficient for continuous streaming** due to computational cost.

**Implementation for real-time:**

* Use a **sliding window** (e.g., last 24 hours or 96 intervals) to feed **Isolation Forest** continuously.
* Label points as **anomalous spikes or drops** if they fall below/above a threshold.


## ** Short-Term Forecasting**

Your goal: **predict energy demand for the next few 15-min intervals**.

### **Model Choice**

1. **SARIMAX**  Recommended

   * Handles **trend + seasonality** (daily & weekly patterns).
   * Can include **exogenous variables** (temperature, weather, holidays) for better accuracy.
2. **Alternative Models**

   * **ARIMA**: Only if no strong seasonal pattern exists.
   * **SARIMA**: If seasonality exists but no exogenous variables are considered.

**Implementation Steps:**

* Resample the data to **15-min frequency**.
* Decompose the time series to inspect **trend and seasonality**.
* Fit **SARIMAX** with exogenous features like `temperature` and `humidity`.
* Forecast **short-term (next 1–4 hours)** for operational decisions.


## ** Validation & Monitoring**

### **Offline Validation**

* Split historical data: **train/test** or **rolling forecast validation**.
* Metrics:

  * **Forecasting:** MAE, RMSE, MAPE
  * **Anomaly Detection:** Precision, Recall, F1-score (if labeled anomalies exist)

### **Online Monitoring**

* Track **real-time forecast errors**:

  * Compare predicted vs actual energy usage.
  * Set alert thresholds for deviations.
* Monitor **anomaly detection drift**:

  * Update the Isolation Forest periodically to adapt to seasonal changes.
* Visual dashboards:

  * Plot forecast vs actual.
  * Show detected anomalies on a timeline.


## **4️ Business Value**

Implementing this solution provides **operational and strategic benefits**:

| Area                    | Impact                                                                      |
| ----------------------- | --------------------------------------------------------------------------- |
| **Grid Stability**      | Detect spikes/drops in real-time → prevent overloads or blackouts.          |
| **Demand Forecasting**  | Accurate short-term forecasts → optimize generation and load balancing.     |
| **Resource Planning**   | Schedule maintenance or energy storage deployment based on predicted usage. |
| **Cost Efficiency**     | Reduce excess energy production and peak-load penalties.                    |
| **Alerts & Automation** | Trigger automated alerts for abnormal events → faster operational response. |

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

np.random.seed(42)
periods = 96
timestamps = pd.date_range(start='2025-01-01', periods=periods, freq='15T')
energy_usage = np.random.normal(loc=50, scale=5, size=periods)

energy_usage[20] = 100
energy_usage[70] = 10

stream_data = pd.DataFrame({'timestamp': timestamps, 'energy_usage': energy_usage})

iso_model = IsolationForest(contamination=0.05, random_state=42)

ts = stream_data.set_index('timestamp')['energy_usage']
sarimax_model = SARIMAX(ts, order=(1,0,1), seasonal_order=(1,1,1,96))
sarimax_fit = sarimax_model.fit(disp=False)

window_size = 24
anomalies_detected = []

for i in range(len(stream_data)):
    current_window = stream_data['energy_usage'].iloc[max(0, i-window_size+1):i+1].values.reshape(-1,1)


    iso_model.fit(current_window)
    is_anomaly = iso_model.predict(current_window[-1].reshape(1,-1))[0] == -1

    if is_anomaly:
        anomalies_detected.append((stream_data['timestamp'].iloc[i], stream_data['energy_usage'].iloc[i]))
        print(f"[ALERT] Anomaly detected at {stream_data['timestamp'].iloc[i]} with usage {stream_data['energy_usage'].iloc[i]:.2f}")

    forecast = sarimax_fit.get_forecast(steps=4)
    forecast_mean = forecast.predicted_mean.values
    threshold = ts.mean() + 3*ts.std()

    if any(forecast_mean > threshold):
        print(f"[ALERT] Forecasted high energy demand at {stream_data['timestamp'].iloc[i]} - next hour may exceed {threshold:.2f}")

plt.figure(figsize=(12,5))
plt.plot(stream_data['timestamp'], stream_data['energy_usage'], label='Energy Usage')
if anomalies_detected:
    anomaly_times, anomaly_values = zip(*anomalies_detected)
    plt.scatter(anomaly_times, anomaly_values, color='red', label='Anomalies', s=100)
plt.xlabel('Timestamp')
plt.ylabel('Energy Usage')
plt.title('Real-Time Energy Usage & Detected Anomalies')
plt.legend()
plt.show()
```
